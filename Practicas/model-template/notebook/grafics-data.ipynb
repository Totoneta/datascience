```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../my-data-1_car_pricing.csv")
df
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="yearOfRegistration",
    y="price"
)
plt.title("Precio y Año de Registro")
plt.xlabel("Año de Registro")
plt.ylabel("Precio")
plt.show()
```

```python
plt.figure()
sns.regplot(
    data=df,
    x='kilometer',
    y='price',
    line_kws={'color':'red'}
)
plt.title('Precio y Kilometraje')
plt.xlabel('Kilometraje')
plt.ylabel('Precio')
plt.show()
```

```python
top_brands = df['brand'].value_counts().head(10).index
df_brands = df[df['brand'].isin(top_brands)]
plt.figure(figsize=(10,5))
sns.boxplot(
    data=df_brands,
    x='brand',
    y='price'
)

plt.title('Distribución de precios según marca')
plt.xlabel('Marca')
plt.ylabel('Precio')
plt.show()
```
