```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/2_discovery_1_car_pricing.csv")
df
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    data=df,
    x='brand',
    y='price'
)

plt.title('Distribución de precios según marca')
plt.xlabel('Marca')
plt.ylabel('Precio')
plt.show()
```
