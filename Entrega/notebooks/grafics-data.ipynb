```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/2_discovery_1_car_pricing.csv")
df
```

# COMENTARIOS SOBRE LOS GRAFICOS EN "Proyecto-Data-Science.odt" PÁGINA 3 Y 4

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="yearOfRegistration",
    y="price"
)
plt.title("Precio y Año de Registro")
plt.xlabel("Año de Registro")
plt.ylabel("Precio")
plt.show()
```

```python
plt.figure()
sns.regplot(
    data=df,
    x='kilometer',
    y='price',
    line_kws={'color':'red'}
)
plt.title('Precio y Kilometraje')
plt.xlabel('Kilometraje')
plt.ylabel('Precio')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    data=df,
    x='brand',
    y='price'
)

plt.title('Distribución de precios según marca')
plt.xlabel('Marca')
plt.ylabel('Precio')
plt.show()
```
