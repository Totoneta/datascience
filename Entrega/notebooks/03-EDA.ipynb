```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/2_discovery_1_car_pricing.csv")
df
```

# Dispersión: Kilómetros y Fecha de Registro

El gráfico de dispersión muestra la relación entre el año de registro del vehículo y su precio. Se observa una tendencia general en la que los vehículos con años de registro más recientes tienden a presentar precios más elevados, mientras que los vehículos más antiguos concentran precios más bajos. No obstante, la dispersión de los puntos indica que, aunque el año de registro influye en el precio, no es el único factor determinante. Existen vehículos relativamente antiguos con precios elevados y vehículos recientes con precios más bajos, lo que sugiere la influencia de otras variables como la marca, el kilometraje y el estado general del vehículo.

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="yearOfRegistration",
    y="price"
)
plt.title("Precio y Año de Registro")
plt.xlabel("Año de Registro")
plt.ylabel("Precio")
plt.show()
```

# Regresión: Kilómetros y Precio

El gráfico de regresión permite identificar la tendencia promedio entre el kilometraje y el precio. La pendiente negativa de la línea indica que, en promedio, a medida que aumenta el kilometraje, el precio del vehículo disminuye. No obstante, la dispersión alrededor de la línea de regresión evidencia una variabilidad considerable, lo que refuerza la idea de que el kilometraje explica solo parcialmente el comportamiento del precio y que otras variables deben ser consideradas para un análisis más completo.

```python
plt.figure()
sns.regplot(
    data=df,
    x='kilometer',
    y='price',
    line_kws={'color':'red'}
)
plt.title('Precio y Kilometraje')
plt.xlabel('Kilometraje')
plt.ylabel('Precio')
plt.show()
```

# Boxplot: Kilómetros y Precio

El boxplot permite comparar la distribución de precios entre las principales marcas. Se observan diferencias claras tanto en los valores centrales como en la dispersión de los precios. Algunas marcas presentan medianas más altas y una mayor variabilidad, mientras que otras muestran precios más concentrados en rangos bajos. La presencia de valores atípicos en varias marcas indica la existencia de modelos con precios significativamente superiores al promedio, lo que confirma que la marca es un factor clave en la determinación del precio de los vehículos.

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    data=df,
    x='brand',
    y='price'
)

plt.title('Distribución de precios según marca')
plt.xlabel('Marca')
plt.ylabel('Precio')
plt.show()
```
